```python
#Jeffrey Yan
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim import corpora
from gensim.models.ldamodel import LdaModel
```

```python
#Read reviews
f=open(r"ratemd.25k.all.txt",'r',encoding='utf-8')
input = f.readlines()
#output format:
# 0 reviewtext
# 1 reviewtext
stop = set(stopwords.words("english"))
#add "doctor" and "dr" to stopwords
stop.update(["doctor","dr"])
#Female doctor reviews (positive, negative)
C11, C12 = list(), list()
#Male doctor reviews (positive, negative)
C21, C22 = list(), list()

#Track current gender of physician
curr_gender= ""

for line in input:
    curr_line = line
    if(curr_line[0]=='D'):
        if(curr_line.find("Female") == -1):
            curr_gender = "Male"
        else:
            curr_gender = "Female"
    else:
        curr_collection = list() #Collection to save review to
        startlen = len("Overall rating: ")
        curr_line=curr_line[startlen:] #remove "Overall rating: "
        ratingstr=curr_line.split(' ',1)[0]
        rating=float(ratingstr)
        if(rating>3.0):
            #Add to female_pos
            if curr_gender == "Female":
                curr_collection = C11
            #Add to male_pos
            else:
                curr_collection = C21
        else:
            #Add to female_neg
            if curr_gender == "Female":
                curr_collection = C12
            #Add to male_neg
            else:
                curr_collection = C22

        #Preprocess
        curr_line = curr_line[len(ratingstr):]#remove rating from string
        curr_line=curr_line.lstrip()#remove leading whitespaces
        #Remove punctuation, numbers and lowercase string
        curr_line = re.sub(r"[^a-zA-z\s]",'',curr_line)
        curr_line = curr_line.lower()
        #Remove stopwords
        curr_line = curr_line.split()
        curr_line = [word for word in curr_line if not word in stop]
        curr_line = ' '.join(curr_line)
        curr_collection.append(curr_line)

#Check that size of collections equal results from Part 1:
# print(len(C11))
# print(len(C12))
# print(len(C21))
# print(len(C22))
```

```python
ccLDA_file = "input_docs.txt"
f = open(ccLDA_file, "w")
all_reviews = list()
#Format lines for ccLDA program:
for rev in C11:
    curr_line = "0 "+rev+"\n"
    all_reviews.append(curr_line)
for rev in C12:
    curr_line = "1 "+rev+"\n"
    all_reviews.append(curr_line)
for rev in C21:
    curr_line = "2 "+rev+"\n"
    all_reviews.append(curr_line)
for rev in C22:
    curr_line = "3 "+rev+"\n"
    all_reviews.append(curr_line)
f.writelines(all_reviews)
    
    
```
