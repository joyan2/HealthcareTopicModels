```python
f=open(r"ratemd.25k.all.txt",'r',encoding='utf-8')
input = f.readlines()
gender_ratings = {
    "Male" : [0,0], 
    #[negative rating count,positive rating count]
    "Female": [0,0]
}
review_length = 0
review_count = 0
min_review = 9999
max_review = 0
curr_gender=""
for line in input:
    curr_line = line
    if(curr_line[0]=='D'):
        if(curr_line.find("Female") == -1):
            curr_gender = "Male"
        else:
            curr_gender = "Female"
    else:
        startlen = len("Overall rating: ")
        curr_line=curr_line[startlen:] #remove "Overall rating: "
        ratingstr=curr_line.split(' ',1)[0]
        rating=float(ratingstr)
        if(rating>3.0):
            gender_ratings[curr_gender][1]+=1
        else:
            gender_ratings[curr_gender][0]+=1
        curr_line = curr_line[len(ratingstr):]#remove rating from string
        curr_line=curr_line.lstrip()#remove leading whitespaces
        currlen = len(curr_line)
        if(currlen>max_review):
            max_review=currlen
        if(currlen<min_review):
            min_review=currlen
        review_length+=currlen
        review_count+=1
#Get review count and % for gender
male_totalcount = gender_ratings["Male"][0]+gender_ratings["Male"][1]
female_totalcount = gender_ratings["Female"][0]+gender_ratings["Female"][1]
male_negativecount = gender_ratings["Male"][0]
male_positivecount = gender_ratings["Male"][1]
female_negativecount = gender_ratings["Female"][0]
female_positivecount = gender_ratings["Female"][1]
#percent:
male_negativepercent=male_negativecount/male_totalcount*100
male_positivepercent=male_positivecount/male_totalcount*100
female_negativepercent=female_negativecount/female_totalcount*100
female_positivepercent=female_positivecount/female_totalcount*100

avg_review_length = review_length/review_count
print("Average review length: " + str(avg_review_length))
print("Min review length: " +str(min_review))
print("Max review length: " +str(max_review))
print("Female doctors' positive review count: " + str(female_positivecount)
      +" Percent: " + str(female_positivepercent))
print("Negative review count: " +str(female_negativecount)
      +" Percent: " + str(female_negativepercent))
print("Total count: " + str(female_totalcount) +" Percent: 100")
print("Male doctors' positive review count: " + str(male_positivecount)
      +" Percent: " + str(male_positivepercent))
print("Negative review count: " +str(male_negativecount)
      +" Percent: " + str(male_negativepercent))
print("Total count: " + str(male_totalcount) +" Percent: 100")






        
```
